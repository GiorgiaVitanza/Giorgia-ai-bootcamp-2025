```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

Leggo i dati dal file csv usando pandas:

```python
with open('HollywoodMovies.csv') as fd:
    data = pd.read_csv(fd) 
data.info()
```

```python
data.describe()
```

Prima analisi dati: film suddivisi per genere e rappresentati con un diagramma a torta

```python
#riempio i NaN con NoGenre
data['Genre'] = data['Genre'].fillna('NoGenre')
labels = list(set(data['Genre']))
sizes = list(data['Genre'].value_counts())
   
```

```python
plt.pie(sizes,labels = labels)
plt.show()
```

Seconda analisi dati: film suddivisi per anno e rappresentati con un istogramma di seaborn

```python
sns.histplot(data['Year'],bins=20,kde=True)
plt.title("Year Distribution of Movies")
plt.show()
```

Terza analisi dati: film raggruppati per AudienceScore ogni anno e rappresentati con barplot di seaborn

```python
sns.barplot(x="Year", y="AudienceScore", data=data)
plt.title("AudienceScore Rate by Year")
plt.show()
```
